```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dark_mode = cv2.imread("horse.jpg") #RGB format image
show_horse = cv2.cvtColor(dark_mode,cv2.COLOR_BGR2RGB) #BGR format image

rainbow = cv2.imread("rainbow.jpg")  #RGB format image
show_rainbow = cv2.cvtColor(rainbow,cv2.COLOR_BGR2RGB) #BGR format image

blue_bricks = cv2.imread("bricks.jpg")  #RGB format image
show_bricks = cv2.cvtColor(blue_bricks,cv2.COLOR_BGR2RGB) #BGR format image
```

# Blue Bricks

```python
img = blue_bricks
color =('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
plt.title('Blue Bricks Image')
plt.show()
```

# Dark Horse

```python
img = dark_mode
color =('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
    plt.ylim(0,500000)
plt.title('Dark Horse Image')
plt.show()
```

# Rainbow

```python
img = rainbow
color =('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
plt.title('Rainbow Image')
plt.show()
```

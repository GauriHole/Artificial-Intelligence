```python
face_cascade = cv2. CascadeClassifier("C:\2-compVision\haarcascade_frontalface_default.xml")
```

```python
def detect_face(img):
    face_img = img.copy()
    face_rects = face_cascade.detectMultiScale(face_img)
    
    for(x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h), (255,255,255),10)
    return face_img
```

```python
result = detect_face(dennis)
```
